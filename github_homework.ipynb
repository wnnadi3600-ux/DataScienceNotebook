1. One purpose of GitHub

GitHub allows you to store and organize your project files online so you can access them anywhere and keep a clean history of your work.

2. What is version control ?

Version control is a system that tracks changes made to files over time. It helps you save different versions of your work, go back if you make mistakes, and avoid losing progress. It makes projects more organized and safer.

3. How GitHub helps in a group project

GitHub makes collaboration easier by allowing multiple people to work on the same project without overwriting each other’s work. Group members can upload updates, review each other’s changes, and merge contributions smoothly.

4. Why employers look at a candidate’s GitHub

Employers review GitHub profiles to see real examples of your coding skills, problem-solving ability, consistency, and project quality. A strong GitHub shows professionalism and practical experience beyond classes.
